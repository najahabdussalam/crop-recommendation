```python
df_fert = df_fert.drop(['Unnamed: 0'],axis=1)
df_fert.head()
```

```python
crop_labels=df_crop['label'].unique()
fert_labels=df_fert['Crop'].unique()
print(fert_labels)
print(crop_labels)
len(crop_labels),len(fert_labels)
```

```python
df_crop['label']=df_crop['label'].astype('str')
df_fert['Crop']=df_fert['Crop'].astype('str')
```

```python
def applychanges(i):
    i=i.replace('  ','')
    i=i.lower()
    return i
```

```python
df_fert['Crop']=df_fert['Crop'].apply(lambda x : applychanges(x))
df_crop['label']=df_crop['label'].apply(lambda x : applychanges(x))
```

```python
crop_labels=df_crop['label'].unique()
fert_labels=df_fert['Crop'].unique()
print(crop_labels)
print(fert_labels)
```

```python
df_fert['Crop']=df_fert['Crop'].replace('mungbeans','mungbean')
df_fert['Crop']=df_fert['Crop'].replace('lentil(masoordal)','lentil')
df_fert['Crop']=df_fert['Crop'].replace('pigeonpeas(toordal)','pigeonpeas')
df_fert['Crop']=df_fert['Crop'].replace('mothbeans(matki)','mothbeans')
df_fert['Crop']=df_fert['Crop'].replace('chickpeas(channa)','chickpea')
```

```python
crop_labels=df_crop['label'].unique()
fert_labels=df_fert['Crop'].unique()
```

```python
final_list=[]
for i in crop_labels:
        if i in fert_labels:
            final_list.append(i)
print(final_list)
```
